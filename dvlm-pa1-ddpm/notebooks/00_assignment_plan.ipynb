## DDPM pipeline in plain English

A diffusion model works in two directions:

### Forward process
We start with a clean image and gradually add small amounts of Gaussian noise over many timesteps until the image becomes almost pure noise.

### Reverse process
We train a neural network to predict the noise present at each timestep. Once the network learns this, we can start from pure noise and repeatedly denoise step by step to generate a sample.

### Why this works
The forward process is fixed and mathematically tractable.
The reverse process is learned using supervised noise prediction.
